```python
import gradio as gr
import fitz  # pymupdf for PDF reading
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
def summarize_paper(file):
    doc = fitz.open(file.name)
    text = ""
    for page in doc:
        text += page.get_text()

    text = text[:1024]  # Limit for BART model

    try:
        summary = summarizer(text, max_length=150, min_length=30, do_sample=False)[0]["summary_text"]
    except Exception as e:
        summary = f"Error during summarization: {e}"

    return summary
```

```python
interface = gr.Interface(
    fn=summarize_paper,
    inputs=gr.File(label="Upload a Research Paper (PDF)"),
    outputs=gr.Textbox(label="AI Summary"),
    title="📄 AI Research Paper Summarizer by DoodleEd",
    description="Upload a research paper and let AI summarize the key findings!"
)
```

```python
interface.launch(share=True)
```
